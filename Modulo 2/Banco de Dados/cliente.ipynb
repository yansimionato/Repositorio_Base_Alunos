```python
import sqlite3
```

```python
cliente = 'clientes.db'
```

# Criar tabelas

```python
script_clientes = '''CREATE TABLE IF NOT EXISTS Clientes(
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                nome TEXT NOT NULL,
                idade INTERGER NOT NULL,
                email TEXT NOT NULL
);'''

try:
    with sqlite3.connect(cliente) as con:
        # criar um cursor
        cur = con.cursor()

        # Executar o script
        cur.execute(script_clientes)

        # Salvar as alterações no banco de dados
        con.commit()
        print("Tabela criada com sucesso!")
except sqlite3.OperationalError as e:
    print('Erro', e)
```

# Consultar tabelas


```python
sql = 'SELECT * FROM Clientes'

try:
    with sqlite3.connect(cliente) as con:
        # Criar um cursor
        cur = con.cursor()

        # Executar o script
        cur.execute(script_clientes)
        res = cur.fetchall() # retorna uma lista de listas
        print(res)
except sqlite3.OperationalError as e:
    print('Erro', e)
```

# Inserir dados


```python
sql = 'INSERT INTO Clientes(nome, idade, email) VALUES (?,?,?)'
try:
    with sqlite3.connect(cliente) as con:
        # Criar um cursor
        cur = con.cursor()

        # Executar o script
        cur.execute(sql, ('Yan', 16, 'yanchoumansimionato@gmail.com'))
        # Salvar as alterações no banco de dados
        con.commit()
except sqlite3.OperationalError as e:
    print('Erro:', e)
```

# Atualizar um registro


```python
sql = 'UPDATE Clientes SET idade = ? WHERE id = ?'

try:
   with sqlite3.connect(cliente) as con:
        # Criar um cursor
        cur = con.cursor()
        # Executar o script
        cur.execute(sql,(18,1))
        # Salvar as alterações no banco de dados
        con.commit()
except sqlite3.OperationalError as e:
    print('Erro:', e)
    
```

# Deletar um registro

```python
sql = 'DELETE FROM Clientes WHERE id = ?'
try:
    with sqlite3.connect(cliente) as con:
        cur = con.cursor()
        cur.execute(sql,(2,))
        con.commit()
except sqlite3.OperationalError as e:
    print('Erro', e)
```

```python
con.close()
print("Conexão encerrada.")
```
