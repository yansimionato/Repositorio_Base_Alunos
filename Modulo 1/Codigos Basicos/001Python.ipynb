# Tutorial Variáveis


## O que são variáveis ?

Variavéis são espaços na memória do computador onde podemos guardar informações ( como textos, números, resultados de cálculos, etc ) para usar depois.

Variável com texto (string)

```python
nome = "Yan"
print(nome)
```

```python
nome = "Pedro"
```

```python
print(nome)
```

Variável com número inteiro(int)

```python
idade = 16
print(idade)
```

```python
# este é um exemplo de número inteiro negativo
temperatura = -3 # esta variável está armazenando um número negativo
print(temperatura) # print() é uma função de saída
```

Variável com número decimal (float)

```python
altura = 1.62 # em python os número decimais são # escritos com  ".", exemplo= 5.2
print(altura)
```

```python
pi = 3.14
print(pi)
```

Variável com valor booleano (bool)

```python
estudante = True
print(estudante)
```

## O que é snake_case?

snake_case é um estilo de escrita usado para nome de variáveis, funções, e nomes em geral no código. Neste estilo, as palavras são todas em minúsculas e separadas por um underline(_)

## Por que usar snake_case?

*   Deixa o código mais fácil de ler
*   É o padrão recomendado pela comunidade Python
*   Segue a PEF8, que é o guia de estilo oficial do Python

OBS: PEF8 é um documento que recomenda boas práticas pra escrever código Python limpo e legível.



## Exemplos que não usam snake_case

```python
nomeCompleto = "Maria" # estilo camelCase (mais comum em JavaScript)
NomeCompleto = "Maria" # estilo PascalCase (usada para classes em Python)
nomecompleto = "Maria" # estilo difícil de ler
```
