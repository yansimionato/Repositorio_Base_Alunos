# Operadores Lógicos

- AND (e)
- OR (ou)
- NOT (negação)

# AND

True + True = True   
True + False = False   
False + True = False   
False + False = False   
   



```python
(10>20) and (20>30) # False + False = False
```

```python
(10<20) and (20<30) # True + True = True
```

# OR

True + True = True  
True + False = True  
False + True = True   
False + False = False     

```python
(1 < 2) or (3 > 6)
```

```python
'João' == 'João' or 'Python' != 'python'
```

# Not

not True = False  
not False = True  

```python
not True
```

```python
not False
```
